## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the subtotal for each line using the unit_price and the qty

df['line_subtotal'] = df['unit_price'] * df['qty']
df[['unit_price', 'qty', 'line_subtotal']].head(3)   #'client_id', 'order_id'
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound for orders over 50 pounds and $10 per pound for items 50 pounds or under.

def shipping_rate(row):
    if (row["total_weight"] > 50):
        return(7)
    else:
        return(10)
# apply to dataframe

df['total_weight'] = df['unit_weight'] * df['qty']
df['shipping_rate'] = df.apply(shipping_rate, axis = 1) 
df['shipping_price'] = df['total_weight'] * df['shipping_rate']
df[['unit_price','unit_weight', 'qty', 'total_weight', 'shipping_price']].head(3)  # 'client_id', 'order_id'
# 'unit_weight',  'shipping_rate',  'line_subtotal', 
```

```python
# Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25%
df['tax'] = df['line_subtotal'] * .0925
df['line_price'] = round(df['line_subtotal'] + df['shipping_price'] + df['tax'], 2)


df[['line_subtotal', 'shipping_price', 'line_price']].head(3)
#'client_id', 'order_id','unit_price', 'qty',  'unit_weight',  'total_weight', 'shipping_rate' , 'tax', 
```

```python
# Create a column for the cost of each line using unit cost, qty, and
# shipping price (assume the shipping cost is exactly what is charged to the client).

# df_edited = df[['first', 'last', 'job', 'phone', 'email', 'client_id', 'order_id', 'order_date', 'order_week', 'order_year', unit_price', 'qty',  'line_subtotal', 'unit_weight',  'total_weight', 'shipping_rate' , 'tax', 'shipping_price', 'total_price']].head(3)

df['line_cost'] = df['unit_cost'] * df['qty']

df = df[['first',
 'last',
 'job',
 'phone',
 'email',
 'client_id',
 'order_id',
 'order_date',
 'order_week',
 'order_year',
 'item_id',
 'category',
 'subcategory',
 'unit_price',
 'unit_cost',
 'unit_weight',
 'qty',
 'line_number',
 'line_subtotal',
 'total_weight',
 'shipping_price',
 'line_price',
 'line_cost'
 ]]


df.head(3)
```

```python
# Create a column for the profit of each line using line cost and line price

df['line_profit'] = df['line_price'] - df['line_cost']
df.head(3)
```
